```python
X = new_car_data.drop('price', axis = 1)
y = new_car_data['price']
```

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size = 0.2, random_state = 0)
```

```python
from sklearn.tree import DecisionTreeRegressor
from sklearn.model_selection import GridSearchCV
```

```python
dtree = DecisionTreeRegressor()
```

```python
param_grid ={
    'max_depth' : [2, 4, 6, 8],
    'min_samples_split' : [2, 4, 6, 8],
    'min_samples_leaf' : [1, 2, 3, 4],
    'max_features' : ['auto', 'sqrt', 'log2'],
    'random_state' : [0, 42]
}
```

```python
grid_search = GridSearchCV (dtree, param_grid, cv=5)
```

```python
grid_search.fit(X_train, y_train)
```

```python
print(grid_search.best_params_)
```

```python
dtree = DecisionTreeRegressor(random_state=0, max_depth=8, max_features='auto',
                             min_samples_split=4, min_samples_leaf=1)
dtree.fit(X_train, y_train)
```

```python
y_pred = dtree.predict(X_test)
```

```python
from sklearn import metrics
import math
```

```python
mae = metrics.mean_absolute_error (y_test, y_pred)
mape = metrics.mean_absolute_percentage_error (y_test, y_pred)
mse = metrics.mean_squared_error (y_test, y_pred)
r2 = metrics.r2_score (y_test, y_pred)
rmse = math.sqrt(mse)
```

```python
print('MAE is {}'.format(mae))
print('MAPE is {}'.format(mape))
print('MSE is {}'.format(mse))
print('R2 Score is {}'.format(r2))
print('RMSE Score is {}'.format(rmse))
```

```python
import_data = pd.DataFrame({
    'Feature Name' : X_train.columns,
    'Importance' : dtree.feature_importances_
})
```

```python
feature = import_data.sort_values(by = 'Importance', ascending = False)

feature_top10 = feature.head(10)
```

```python
plt.figure(figsize=(10,8))
sns.barplot(data=feature_top10, x='Importance', y='Feature Name')
plt.title('Feature Importance Attribute (Decision Tree Regressor)', fontsize=15)
plt.xlabel('Importance', fontsize=12)
plt.ylabel('Feature Name', fontsize=12)
plt.show()
```
